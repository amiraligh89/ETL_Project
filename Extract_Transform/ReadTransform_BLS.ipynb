```python
# Dependencies
import pandas as pd
import os
import csv
import glob

inpath = os.path.join('.','Input_Data')
outpath = os.path.join('..','Load')
print(inpath)
```

```python
## Reading in files from the BLS website for 12 retail industries##
all_files = glob.glob(os.path.join(inpath,"*.xlsx"))

#print(all_files)

li = []

for filename in all_files:
    print(f"File Name: {filename}")
    
    # substring the input path out of the file name
    file_only = filename[len(inpath)+1:]

    print(f"Spreadsheet Name: {file_only}")
    print("------")

    df = pd.read_excel(filename, skiprows=12)
    df['Filename'] = file_only 
    df['NAICS_Code'] = file_only[0:3]
    li.append(df)

frame = pd.concat(li, axis=0, ignore_index=True)
frame
```

```python
## Rearranging the columns to make NAICS_Code & Year as primary index columns 
frame = frame[['NAICS_Code','Year','Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec', ]]
frame.head()
```

```python
# determine data types
frame.dtypes
```

```python
# Using the melt function to massage data set into a vertical format sorted by NAICS_Code and Year ##
emp_data = frame.melt(id_vars=["NAICS_Code","Year"],var_name=["Month"],value_name="Num_Employees")
emp_data.sort_values(by=['Year','Month'])
emp_data.head()
```

```python
# Export file as a CSV, without the Pandas index, but with the header
outfile = os.path.join(outpath,'final_emp_data.csv')
emp_data.to_csv(outfile, index=False, header=True)
```
